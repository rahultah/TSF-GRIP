## Import all the necessary Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

## Import Dataset

```python
df=pd.read_csv("http://bit.ly/w-data")
```

## Plot the graph using the CSV file

```python
%matplotlib inline
plt.xlabel("Hours studied")
plt.ylabel("Scores")
plt.scatter(df.Hours,df.Scores,color="Green",marker="*")
plt.show()
```

```python
x = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                            test_size=0.2, random_state=0) 
```

## Training   

```python


regr = LinearRegression()
regr.fit(x_train, y_train)


```

## We know that the equation on Simple regression line is =  y=mx+c

```python
line_eq = regr.coef_* x +regr.intercept_ #y = mx+c
```

## Plotting the Line 

```python
plt.scatter(x,y)
plt.plot(x,line_eq)
plt.show()
```

## The below values in the x_test are the ones which we are going to test to make predictions

```python
print(x_test)
```

```python
pred_val = regr.predict(x_test)
```

```python
df1 = pd.DataFrame({"Actual Values: -":y_test, "Predicted Values: - ":pred_val})
df1
```
